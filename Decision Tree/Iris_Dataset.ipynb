```python
##import seaborn to use its dataset of iris
import seaborn as sns
```

```python
##load the dataset into dataframe df and print its head
df = sns.load_dataset('iris')
df.head()
```

```python
##independent features and dependent features of iris dataset assigned to X and y respectively
X = df.iloc[:,:-1]
y = iris.target
```

```python
##print X and y to verify if independent and dependent variables are seperated
X, y
```

```python
##import train test split from sklearn
from sklearn.model_selection import train_test_split
```

```python
##divide the dataset into train dataset and test dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
##import DecisionTreeClassifier from sklearn
from sklearn.tree import DecisionTreeClassifier
```

```python
##Postpruning and the default criterion is gini so we changed it to entropy
treemodel = DecisionTreeClassifier(criterion="entropy", max_depth = 2)
```

```python
##fit the model with our training dataset
treemodel.fit(X_train, y_train)
```

```python
##import tree from sklearn
from sklearn import tree
```

```python
##construct a decision tree 
plt.figure(figsize = (15,10))
tree.plot_tree(treemodel,filled = True)
```

```python
#prediction of our test dataset
y_pred = treemodel.predict(X_test)
y_pred
```

```python
##import accuracy_score to measure how accurate our model is
from sklearn.metrics import accuracy_score, classification_report
```

```python
##print the score using accuracy_score
score = accuracy_score(y_pred , y_test)
print(score)
```

```python
##use classification_report for more detailed report whict provides different scores including f1 score.
print(classification_report(y_pred, y_test))
```
