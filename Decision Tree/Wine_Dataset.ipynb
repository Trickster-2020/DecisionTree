## Import Wine dataset from sklearn datasets and load into 'data'

```python
from sklearn.datasets import load_wine
data = load_wine()
```

# Display wine dataset stored in data

```python
data
```

# Load attributes of wine data into 'x'

```python
x = data.data
x
```

# Load target/lable of wine data into 'y'

```python
y = data.target
y
```

# Import train_test_split from sklearn to divide the data into train and test datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# Import DecisionTreeClassifier from sklearn for our classification. Here criterion can either be 'gini' by default, 'entropy' or 'log_loss'.

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini = DecisionTreeClassifier(random_state=0)
```

# Train the model using training datasets. Here we are using gini.

```python
clf_g = clf_gini.fit(X_train, y_train)
```

# Import tree for plotting the tree and also matplotlib for defining the size of tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

# Define figsize and plot the tree using plot_tree

```python
plt.figure(figsize = (15,10))
tree.plot_tree(clf_g,filled = True, rounded = True)
```

# We repeat same steps but this time using criterion as 'entropy'

```python
clf_entropy = DecisionTreeClassifier(random_state=0, criterion='entropy')
clf_e = clf_entropy.fit(X_train, y_train)
```

```python
plt.figure(figsize = (15,10))
tree.plot_tree(clf_e,filled = True, rounded = True)
```

# Predict the target using gini 

```python
y_pred = clf_g.predict(X_test)
y_pred
```

# Import accuracy_score and compute accuracy for gini 

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

# Predict the target using entropy

```python
y_pred_entropy = clf_entropy.predict(X_test)
y_pred_entropy
```

# Compute accuracy for entropy

```python
accuracy_score(y_test, y_pred_entropy)
```
